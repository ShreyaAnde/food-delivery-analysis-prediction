```python

def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    
    print(f"\nColumn: {column}")
    print(f"Lower Bound: {lower_bound}")
    print(f"Upper Bound: {upper_bound}")
    print(f"Number of Outliers: {len(outliers)}")
    
    return outliers

# Apply to all numeric columns
numeric_columns = ["Distance_km", "Preparation_Time_min", 
                   "Courier_Experience_yrs", "Delivery_Time_min"]

for col in numeric_columns:
    detect_outliers_iqr(df, col)
```

```python
#Does Traffic Level affect Delivery Time?
#H₀: Mean delivery time is same for all traffic levels
#H₁: At least one traffic level has different mean
```

```python
from scipy.stats import f_oneway

group0 = df[df['Traffic_Level']==0]['Delivery_Time_min']
group1 = df[df['Traffic_Level']==1]['Delivery_Time_min']
group2 = df[df['Traffic_Level']==2]['Delivery_Time_min']

f_stat, p_value = f_oneway(group0, group1, group2)

print("P-value:", p_value)
```

```python
#### p > 0.05 → No significant effect
```

#Does Weather affect Delivery Time? chi  square test
H₀: Traffic and Weather are independent
H₁: They are related

```python
import pandas as pd
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(df['Traffic_Level'], df['Weather'])

chi2, p, dof, expected = chi2_contingency(contingency_table)

print("P-value:", p)
```

#### p > 0.05 → No relationship


##### “To statistically validate whether traffic impacts delivery time, I performed an ANOVA test. The p-value was below 0.05, indicating a statistically significant difference in mean delivery time across traffic levels.”

```python
x = df.drop(['Delivery_Time_min', 'Order_ID'], axis=1)

y=df['Delivery_Time_min']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
print("X_train shape before scaling:", X_train.shape)
print("Columns:", X_train.shape)

```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

import joblib
joblib.dump(scaler, "scaler.pkl")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
dt = DecisionTreeRegressor(random_state=42)
dt.fit(X_train, y_train)
```

```python
rf = RandomForestRegressor(n_estimators=100,random_state=42)
rf.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

def evaluate_model(y_test, y_pred, model_name):
   
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)

    print(f"Model: {model_name}")
    print(f"MAE: {mae:.4f}")
    print(f"MSE: {mse:.4f}")
    print(f"RMSE: {rmse:.4f}")
    print(f"R2 Score: {r2:.4f}")
    print("-" * 40)

    return {"MAE": mae, "MSE": mse, "RMSE": rmse, "R2": r2}
```

```python
lr_pred = lr.predict(X_test)
dt_pred = dt.predict(X_test)
rf_pred = rf.predict(X_test)

evaluate_model(y_test, lr_pred, "Linear Regression")
evaluate_model(y_test, dt_pred, "Decision Tree")
evaluate_model(y_test, rf_pred, "Random Forest")
```
